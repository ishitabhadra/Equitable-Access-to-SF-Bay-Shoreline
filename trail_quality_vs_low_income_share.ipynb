```python
import pandas as pd
import matplotlib.pyplot as plt
import os

#Load data
df = pd.read_csv("Shoreline-Access-Pts_v2-1-attribute-table.csv")

#Select columns to use
TRAIL_COL = "mean_Trail_Quality_Score"
LOW_INC_COL = "SUM_Estimated_Low_Income_Households"
HH_COL = "SUM_Estimated_Total_Households"

#Compute low-income share
df["low_income_share"] = df[LOW_INC_COL]/df[HH_COL]

#Create scatter plot
plt.figure(figsize=(8, 5))
plt.scatter(
    df[TRAIL_COL],
    df["low_income_share"],
    alpha=0.6
)

plt.xlabel("Mean Trail Quality Score")
plt.ylabel("Low-Income Household Share")
plt.title("Trail Quality vs Low-Income Household Share")
plt.tight_layout()
plt.savefig("trail_quality_vs_low_income_share.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
#Compute low-income share
df["low_income_share"] = (
    df["SUM_Estimated_Low_Income_Households"] /
    df["SUM_Estimated_Total_Households"]
)

#Compute correlation
correlation = df["mean_Trail_Quality_Score"].corr(df["low_income_share"])

print("Correlation:", correlation)
```
